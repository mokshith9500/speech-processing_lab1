```python
# Extracting a part of audio as per user's choice
import librosa
import librosa.display
from IPython.display import Audio
import matplotlib.pyplot as plt

# Load the audio file
y, sr = librosa.load(r"C:\Users\Sreek\Dropbox\PC\Downloads\LAB_1.wav")

# Define the start and end time (in seconds) for the segment
start_time = int(input("Enter starting second: "))  # Replace with the desired start time
end_time = int(input("Enter ending second: "))      # Replace with the desired end time

# Extract the segment of the signal
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)
segment = y[start_sample:end_sample]

# Play the extracted segment
display(Audio(data=segment, rate=sr))

# Plot the extracted signal
plt.figure(figsize=(12, 4))
librosa.display.waveshow(segment, sr=sr)
plt.title('Extracted Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Reverse the signal
import librosa
from IPython.display import Audio
import numpy as np
y, sr = librosa.load(r"C:\Users\Sreek\Dropbox\PC\Downloads\LAB_1.wav")

# Reverse the audio signal
reversed_y = np.flipud(y)

# Play the reversed audio
display(Audio(data=reversed_y, rate=sr))

# Plot the original and reversed signals
plt.figure(figsize=(12, 6))

# Original signal
plt.subplot(2, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Reversed signal
plt.subplot(2, 1, 2)
librosa.display.waveshow(reversed_y, sr=sr)
plt.title('Reversed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
